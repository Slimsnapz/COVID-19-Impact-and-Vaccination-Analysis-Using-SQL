```python
%%sql

-- Selecting the data I want to work on
-- The data shows the likelihood of dying after contracting COVID-19 in Nigeria
-- Calculating the percentage of deaths from total cases
SELECT location, date, total_cases, total_deaths, (total_deaths / total_cases) * 100 AS deaths_percentage
FROM covid_deaths
WHERE location LIKE '%Nigeria%'
ORDER BY 1, 2
LIMIT 100
```

```python
%%sql


-- Looking at total cases vs the population 
-- Shows the percentage of the population in Nigeria infected with COVID
SELECT location, date, total_cases, population, (total_cases / population) * 100 AS percentage_infected
FROM covid_deaths
WHERE location = 'Nigeria'
LIMIT 100;
```

```python
%%sql

-- Analyzing countries with the highest infection rate compared to their population
-- Calculating the maximum total cases and the percentage of the population infected
SELECT location, population, MAX(total_cases) AS highest_infected_country, MAX(total_cases / population) * 100 AS percentage_infected
FROM covid_deaths
WHERE continent IS NOT NULL
GROUP BY location, population 
ORDER BY highest_infected_country DESC
LIMIT 100
```

## Breaking it down by continent

```python
%%sql

-- Retrieving countries with the highest total death count from COVID-19, considering only those with a known continent
SELECT location, MAX(total_deaths) AS total_death_count
FROM covid_deaths
WHERE continent IS NOT NULL
GROUP BY location
ORDER BY total_death_count DESC
LIMIT 100
```

```python
%%sql

-- Showing the continent with the highest total death count from COVID-19
SELECT continent, MAX(total_deaths) AS total_death_count
FROM covid_deaths
WHERE continent IS NOT NULL
GROUP BY continent
ORDER BY total_death_count DESC
```

## Global Numbers


```python
%%sql

-- Summarizing global COVID-19 data: total new cases and deaths per day
-- Calculating the percentage of deaths from total new cases worldwide
SELECT date,
    SUM(new_cases) AS world_cases,
    SUM(new_deaths) AS deaths_worldwide,
    (SUM(new_deaths) / SUM(new_cases)) * 100 AS percentage_death_worldwide
FROM covid_deaths
WHERE continent IS NOT NULL
GROUP BY date
LIMIT 1000;
```

```python
%%sql

-- Showing total population and the number of newly vaccinated people in Nigeria
-- Calculating the cumulative number of vaccinated individuals over time
SELECT 
    covid_deaths.continent,
    covid_deaths.location, 
    covid_deaths.date, 
    covid_deaths.population, 
    covid_vaccinations.new_vaccinations,
    SUM(covid_vaccinations.new_vaccinations) 
        OVER (PARTITION BY covid_deaths.location 
              ORDER BY covid_deaths.location, covid_deaths.date) AS rolling_vaccinated_people
FROM 
    covid_deaths
JOIN 
    covid_vaccinations
    ON covid_deaths.location = covid_vaccinations.location
    AND covid_deaths.date = covid_vaccinations.date
WHERE 
    covid_deaths.continent IS NOT NULL 
    AND covid_deaths.location = 'Nigeria'
ORDER BY 
    covid_deaths.continent, covid_deaths.location, covid_deaths.date
LIMIT 1000;
```

```python
%%sql

-- Retrieving total population and the cumulative number of vaccinations in each country
SELECT 
    covid_deaths.continent,
    covid_deaths.location, 
    covid_deaths.date, 
    covid_deaths.population, 
    covid_vaccinations.new_vaccinations,
    SUM(covid_vaccinations.new_vaccinations) 
        OVER (PARTITION BY covid_deaths.location 
              ORDER BY covid_deaths.date) AS rolling_vaccinated_people
FROM 
    covid_deaths
JOIN 
    covid_vaccinations
    ON covid_deaths.location = covid_vaccinations.location
    AND covid_deaths.date = covid_vaccinations.date
WHERE 
    covid_deaths.continent IS NOT NULL 
ORDER BY 
    covid_deaths.continent, covid_deaths.location, covid_deaths.date
LIMIT 1000;
```

## Using CTE (common table expression)

```python
%%sql

-- Creating a Common Table Expression (CTE) to analyze vaccination data by continent and location
-- This CTE calculates the rolling total of new vaccinations and retrieves population data
WITH pop_vs_vac (continent, location, date, population, new_vaccinations, rolling_vaccinated_people) AS
(
    SELECT 
        covid_deaths.continent,
        covid_deaths.location, 
        covid_deaths.date, 
        covid_deaths.population, 
        covid_vaccinations.new_vaccinations,
        SUM(covid_vaccinations.new_vaccinations) 
            OVER (PARTITION BY covid_deaths.location 
                  ORDER BY covid_deaths.location, covid_deaths.date) AS rolling_vaccinated_people
    FROM 
        covid_deaths
    JOIN 
        covid_vaccinations
        ON covid_deaths.location = covid_vaccinations.location
        AND covid_deaths.date = covid_vaccinations.date
    WHERE 
        covid_deaths.continent IS NOT NULL 
    ORDER BY 
        covid_deaths.continent, covid_deaths.location, covid_deaths.date
)
SELECT *, (rolling_vaccinated_people / population) * 100 AS percentage_of_vaccinated_people
FROM pop_vs_vac
LIMIT 1000;
```
